```python
from ast import literal_eval
import matplotlib.pyplot as plt

# Paper order: slac, musedata, lakh

with open('slac.py') as f:
    mainlist = [literal_eval(line) for line in f]
    
slac = mainlist[0]

with open('musedata.py') as f:
    mainlist = [literal_eval(line) for line in f]
    
musedata = mainlist[0]

with open('lakh.py') as f:
    mainlist = [literal_eval(line) for line in f]
    
lakh = mainlist[0]


notes_slac = {}
notes_musedata = {}
notes_lakh = {}
for i in range(128):
    notes_slac[i] = slac['note_' + str(i)]
    notes_musedata[i] = musedata['note_' + str(i)]
    notes_lakh[i] = lakh['note_' + str(i)]
    
def average(d):    
    avg = 0
    total = 0
    for key, value in d.items():
        avg += (key * value)
        total += value
    avg = avg / total
    return avg

print(average(notes_slac))
print(average(notes_musedata))
print(average(notes_lakh))
    
plt.bar(notes_slac.keys(), notes_slac.values(), color='r')
plt.show()

plt.bar(notes_musedata.keys(), notes_musedata.values(), color='g')
plt.show()

plt.bar(notes_lakh.keys(), notes_lakh.values(), color='b')
plt.show()

```

```python
# Instruments

instruments_slac = {}
instruments_musedata = {}
instruments_lakh = {}
for i in range(128):
    instruments_slac[i] = slac['program_' + str(i)]
    instruments_musedata[i] = musedata['program_' + str(i)]
    instruments_lakh[i] = lakh['program_' + str(i)]

print(instruments_slac) # 0, 25, 48 - acoustic grand piano, acoustic guitar (steel), string ensemble 1
print(instruments_musedata) # 6, 48 - harpsichord, string ensemble 1
print(instruments_lakh) # 0, 25, 48, 123 - acoustic grand piano, acoustic guitar (steel), string ensemble 1, bird tweet
    
plt.bar(instruments_slac.keys(), instruments_slac.values(), color='r')
plt.show()

plt.bar(instruments_musedata.keys(), instruments_musedata.values(), color='g')
plt.show()

plt.bar(instruments_lakh.keys(), instruments_lakh.values(), color='b')
plt.show()
```
