```python
# Cleaning data frame

def label_dataset(row):
    if 'slac' in row['filename']:
        return 'SLAC'
    if 'musedata' in row['filename']:
        return 'MuseData'
    if 'lakh' in row['filename']:
        return 'Lakh'
    return 'Other'

df.apply(lambda row: label_dataset(row), axis=1)

df_clean = df[['program_change', 'note_on', 'set_tempo', 'time_signature']]
df_clean['dataset'] = df.apply(lambda row: label_dataset(row), axis=1)

df_clean[df_clean['dataset'] == 'Lakh'].sort_values('program_change', ascending=False, na_position='last')

df_clean[df_clean['dataset'] == 'Lakh']
# print(df_clean.loc[df_clean['dataset'] == 'Lakh'])
# print(max(df_clean['note_on']))
# print(max(df_clean['program_change']))

# plt.title('Events per type of event, all datasets')
# plt.xlabel('Event type')
# plt.ylabel('Counts')
# plt.grid(axis='y', alpha=0.75)
```
